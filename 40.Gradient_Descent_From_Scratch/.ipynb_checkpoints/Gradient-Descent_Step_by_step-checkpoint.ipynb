```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
#Lets Apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color="red")
```

### Lets Appply Gradient Descent Assuming Slope is Constant m=78.35<br>Lets Assume the Starting value of b is b=0

```python
y_pred=((78.35*X)+100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

## In the Above Figure you can see that green line is far from the Best Fit Line

```python
m=78.50
b=100

loss_slope=-2*np.sum(y-m*X.ravel()-b)
#using Formula
loss_slope
```

```python
#Lets take learning Rate 0.1
lr=0.1

step_size=loss_slope*lr
step_size
```

```python
# Calculate the new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

#### You can see that in the above iteration green line move closer to red line

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

## Green Line Moves CLoser to red line

```python
#iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((78.35*X)+b).reshape(4)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

## You can see that , in the above Graph Green line is almost fit over the best fit line 

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100
#iterations

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```

## In the above Cell you can see that , we use for loop to use iterations , such that line is moving closely to best fit line slowly slowly
