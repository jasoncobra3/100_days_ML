```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df=pd.read_csv("weight-height.csv")
```

```python
df.head()
```

```python
df["Height"].describe()
```

```python
df.shape
```

```python
plt.figure(figsize=(18,8))

plt.subplot(1,2,1)
sbn.distplot(df["Height"])

plt.subplot(1,2,2)
sbn.boxplot(df["Height"])
```

#### from the above graph You can see the outliers in height Column

```python
upper_limit=df["Height"].quantile(0.99)
upper_limit
```

```python
lower_limit=df["Height"].quantile(0.01)
lower_limit
```

### In the Percentile Method Note that upper_limit and Lower_limit must be in equal proportion ,which mean if your upper_limit is 0.95 then lower_limit must be 0.05

```python
#Finding Outlier
df[(df["Height"]>=74.78)|(df["Height"]<=58.13)]
#below is the outliers in Height Column
```

# Trimming

```python
new_df=df[(df["Height"]<=74.78)&(df["Height"]>=58.13)]
```

```python
new_df
```

### You can see that Shape of new data is reduced which means Outliers are remove successfully

```python
df["Height"].describe()
```

```python
new_df["Height"].describe()
```

#### You can see that in the above info  Only Min and Max value of the data is changed while remain info is reamins same

```python
#comparing
plt.figure(figsize=(18,8))

plt.subplot(2,2,1)
sbn.distplot(df["Height"])

plt.subplot(2,2,2)
sbn.boxplot(df["Height"])

plt.subplot(2,2,3)
sbn.distplot(new_df["Height"])

plt.subplot(2,2,4)
sbn.boxplot(new_df["Height"])
```

#### From the above Graph you can see that Outliers in the Data are removed Successfully while  distribution changes Minor

# Capping(Winsorization)

```python
new_data=df.copy()
new_data["Height"]=np.where(
    df["Height"]>=upper_limit,
    upper_limit,
    np.where(
    df["Height"]<=lower_limit,
        lower_limit,
        df["Height"]
    )

)
```

```python
new_data
```

```python
new_data.shape
```

#### from the above data shape is not change which means Outliers are capped succesfully!

```python
new_data["Height"].describe()
```

##### From the above Max and min values for Height column is changed without change in any other values 

```python
#comparing

plt.figure(figsize=(18,8))

plt.subplot(2,2,1)
sbn.distplot(df["Height"])

plt.subplot(2,2,2)
sbn.boxplot(df["Height"])

plt.subplot(2,2,3)
sbn.distplot(new_data["Height"])

plt.subplot(2,2,4)
sbn.boxplot(new_data["Height"])
```

### From the above graph you can notice that outliers are treted successfully bby capping them on upper limit and lower limit , while in distribution of data you cn see that two columns are raised unexpectedly becuase all outliers are capped and added to these last columns
