```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_data=pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
#usinng only two columns for clustering as Annual income and spending Score
data=customer_data.iloc[:,3:5].values
```

```python
data
```

- Using Scipy for plotting dendograms 

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendogram")
dend=shc.dendrogram(shc.linkage(data,method='ward'))
```

##### - From Above Dendogram you can see that , the vertical line  with no cuts of horizontal line is making 5 clusters , such that we can form 5 clusters for Agglomerative clustering.

```python
from sklearn.cluster import AgglomerativeClustering

cluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
labels=cluster.fit(data)
```

#### From above we can see that number of clusters is 5(n_clusters=5) <br>metrics used for calculating distance between points is euclidean (affinity='euclidean') <br>type of clustering we use is Ward (linkage='ward') 

```python
labels
```

# Plotting graph of clusters

```python
plt.figure(figsize=(10,7))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_,cmap='rainbow')
```

#### From above graph you can see that we can clearly form a 5 clusters using Dendogram and Agglomerative heirarchial Clustering.<br>Such that in case of k-means Clustering it may be difficult to form clusters in this type of data.
