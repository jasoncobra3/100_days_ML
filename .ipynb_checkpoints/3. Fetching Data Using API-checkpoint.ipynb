```python
data=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1")
```

```python
data.json() #Converting data into json
```

```python
data.json()["results"] #Selecting data only on the basis of Data
```

```python
temp_df=pd.DataFrame(data.json()["results"])[["id","original_language","original_title","popularity","release_date","vote_average","vote_count"]]
```

```python
temp_df #At this point we are filtering data from json format to tabular format
```

```python
temp_df.shape
```

```python
df=pd.DataFrame()
```

```python
df
```

```python
for i in range(1, 429):
    # Make the API request
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Extract relevant data from the response
        temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        
        # Append the data to the main DataFrame
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Failed to retrieve data for page {i}")
```

```python
print(df)
```

```python
df.to_csv("movies.csv")
```
