```python
import pandas as pd 
import numpy as  np
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
X_train.shape
```

### For KNN data must be in same scale<br> Hence using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler() 

X_train=std.fit_transform(X_train)
X_test=std.transform(X_test)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred=knn.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

### From above you can see that , model overfits over the data , Since the Value of K is not suitable for model or its minimum.<br>Hence to train the model perfectly and achieve a correct accuracy we need to find the optimum numbe of neighbors that is value of K <br> We can find the value of k using experimentation.

```python
#Performing a experimentation 
score=[]

for i in range(1,16):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    score.append(accuracy_score(y_test,y_pred))
```

```python
score
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,16),score)
```

### From above graph you can see that, for k=3 the accuracy of model is 99 hence this is optimum value of k <br>K must be Even always.

# Tool For Decision Surface of KNN

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.widgets import Slider

def load_data():
    cancer = datasets.load_breast_cancer()
    return cancer

def plot_decision_boundaries(n_neighbors):
    h = .02
    cmap_light = ListedColormap(['orange', 'blue'])
    cmap_bold = ListedColormap(['darkorange', 'darkblue'])

    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(8, 6))
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(f'2-Class classification (k = {n_neighbors})')
    plt.show()

cancer = load_data()

# Use only the first two features and standardize them.
X = StandardScaler().fit_transform(cancer.data[:, :2])
y = cancer.target

# Create a slider to control the number of neighbors
slider_ax = plt.axes([0.1, 0.01, 0.8, 0.03], facecolor='lightgoldenrodyellow')
n_neighbors_slider = Slider(slider_ax, 'Neighbors', 1, 20, valinit=5, valstep=1)

# Define an update function for the slider
def update(val):
    n_neighbors = int(n_neighbors_slider.val)
    plot_decision_boundaries(n_neighbors)

n_neighbors_slider.on_changed(update)

# Plot initial decision boundaries with default value
plot_decision_boundaries(3)
plt.show()
```

- From above Decision Region you can tune the model with number of neighbors i.e K 
- You can see that smaller the value of k , model will Overfit
- For Greater value of k , model will underfit.
- So to achieve accuracy while testing we need optimum value of k, nor small neither large.
