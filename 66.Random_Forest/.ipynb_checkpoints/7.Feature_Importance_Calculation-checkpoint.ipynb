```python
from sklearn.datasets import make_classification
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y=make_classification(n_samples=5,n_classes=2,n_features=2, n_informative=2,n_redundant=0,random_state=0)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

- From above you can see that the above feature importance is same as the calculated feature importance using formula.
- From above we can say that 0th feature is more important than 1th feature.
- 0th feature importance 0.625
- 1th feature imporatnce 0.375

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=2)
rf.fit(X,y)
```

```python
rf.feature_importances_
```

#### Feature importance in case of Random_Forest
- The only differnce is that in case of Decision tree we calculate a feature importance on single decision tree
- In case of Random Forest we take the Average of decision trees feature importance to calculate actual feature importance.


```python
print(rf.estimators_[0].feature_importances_) #For 1st decision tree feature importance
print(rf.estimators_[1].feature_importances_) #For 2nd decision tree feature importnace

print((1 + 0.555)/2)#For 1st column of RandomForest
print((0 + 0.444)/2)#For 2nd column of RandomForest
```

- From above feature imporatance of 1st decision tree 1st column is 0.55555556
- While for feature importance of 2nd descision tree of 2nd column is 
- Thus RandomForest Average out the result fof feature importance of all decision trees.
