# Code

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_features=5,n_redundant=0,n_informative=5,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
print(df.shape)
df.head()
```

### 1.Function for Row Sampling

```python
def sample_rows(df,percent):
    return df.sample(int(percent*df.shape[0]),replace=True)
```

### 2.Function for Column Sampling

```python
import random
def sample_features(df,percent):
  cols = random.sample(df.columns.tolist()[:-1],int(percent*(df.shape[1]-1)))
  new_df = df[cols]
  new_df['target'] = df['target']
  return new_df
```

### 3.Function for combined sampling

```python
def combined_sampling(df,row_percent,column_percent):
    new_df=sample_rows(df,row_percent)
    return sample_features(new_df,column_percent)
```

# 1.Row Sampling

```python
df1=sample_rows(df,0.2)
```

```python
df1.shape
```

```python
df2=sample_rows(df,0.2)
```

```python
df3=sample_rows(df,0.2)
```

```python
df2.shape
```

```python
df3.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf1=DecisionTreeClassifier()
clf2=DecisionTreeClassifier()
clf3=DecisionTreeClassifier()
```

```python

clf1.fit(df1.iloc[:,0:5],df1.iloc[:,-1])
clf2.fit(df2.iloc[:,0:5],df2.iloc[:,-1])
clf3.fit(df3.iloc[:,0:5],df3.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(clf1)
```

```python
plot_tree(clf2)
```

```python
plot_tree(clf3)
```

## From above  tree graphs you can see that , each sample data works differently for each base model using Row_sampling.
