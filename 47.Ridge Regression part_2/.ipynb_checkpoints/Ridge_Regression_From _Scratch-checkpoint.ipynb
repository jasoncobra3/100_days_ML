```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr=Ridge(alpha=0.1,solver="cholesky")
#Cholesky is method to solve Ridge just line OLS(ordinary List Square)
```

```python
rr.fit(X_train,y_train)
```

```python
y_pred=rr.predict(X_test)
print("R2_score:",r2_score(y_test,y_pred))
```

```python
print(rr.coef_)
print(rr.intercept_)
```

```python
#Making Ridge For N dimension
class MeraRidge:
    def __init__(self,alpha=0.1):
        self.coef_=None
        self.intercept_=None
        self.alpha=alpha
        
    def fit(self,X_train,y_train):
        
        X_train=np.insert(X_train,0,1,axis=1)
        I=np.identity(X_train.shape[1])
        I[0][0]=0
        result=np.linalg.inv(np.dot(X_train.T,X_train)+self.alpha*I).dot(X_train.T).dot(y_train)
        self.intercept_=result[0]
        self.coef_=result[1:]
        
    def predict(self,X_test):
        
        return np.dot(X_test,self.coef_)+self.intercept_
        
```

### In the above class we make the First column of matrix Zero (I[0][0]=0), as in Sklearn's Ridge class they make the first column of matriz 0 , so to obtain a accurate answer just as Sklearn's Ridge clas we make this change

```python
reg=MeraRidge(alpha=0.1)
reg.fit(X_train,y_train)
y_predict=reg.predict(X_test)
print("R2_score:",r2_score(y_test,y_predict))
print(reg.coef_)
print(reg.intercept_)
```

## From above you can see that Accuracy_score , Coefficinet and Intercept we calculated using Our Own Class is Same as the values in Sklearn's Ridge Class , Hence we bulild a accurate Class for Ridge regression in N dimension.
