### From Above Graph you can see the drastic change in model For Orange line model is Underfitting while for line with alpha 10 it moves slightly upwards making minimum loses

```python
def linear_regression(X,y,alpha=1):
    x_mean = X.mean()
    y_mean = y.mean()
    
    num = 0
    den = 0
    
    for i in range(X.shape[0]):
        num = num + (y[i] - y_mean) * (X[i] - x_mean)
        den = den + (X[i] - x_mean) * (X[i] - x_mean)
        
    m = num/(den + alpha)
    b = y_mean - m*x_mean
        
    return m,b
```

```python
#Making Our Own Ridge Class
class MeraRidge:
    def __init__(self,alpha=0.1):
        self.m=None
        self.b=None
        self.alpha=alpha
        
    def fit(self,X_train,y_train):
        num=0
        den=0
        
        for i in range(X_train.shape[0]):
            num=num+(y_train[i]-y_train.mean())*(X_train[i]-X_train.mean())
            den=den+(X_train[i]-X_train.mean())*(X_train[i]-X_train.mean())
        self.m=num/(den+self.alpha)
        self.b=y_train.mean()-(self.m*X_train.mean())
        print(self.m,self.b)
        
    def predict(self,X_test):
        return self.m*X_test + self.b
        
```

```python
Ridge=MeraRidge(alpha=100)
```

```python
Ridge.fit(X,y)
```

## From above you can see that the value for alpha=100 Our Ridge Regression class is Same as that of we calculated with the help of Sklearn's Ridge Class , hence Our Class is working Properly with optimum solution
