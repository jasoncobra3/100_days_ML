# Sigmoid Function

### Sigmoid function not only pull the line towards points but it also push line from the points

## In sigmoid function<br> Misclassified points Pull the lines towards themselfs <br> Classified points Push the line from themselfs

## There are total four Conditions<br>1.Misclassified point closer to line pull lines with smaller magnitude<br>2.Misclassified point far from line pull lines with greater magnitude<br>3.Classified point closer to line will push the line with greater magnitude <br> 4.Classified point far from the line will push the line with smaller magnitude.
