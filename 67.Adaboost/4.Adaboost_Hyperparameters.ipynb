class sklearn.ensemble.AdaBoostClassifier(estimator=None, *, n_estimators=50, learning_rate=1.0, algorithm='SAMME.R', random_state=None)

### estimator=None
- The base estimator from which the boosted ensemble is built.
- Estimator stands for the base model use for prediction in AdaBoost
- By Default estimators are Deicision Stumps (Decision tree with height 1)

### n_estimators=50
- This parameters decides the number of decision stumps to be trained in Adaboost.
- Its default value is 50.
- The maximum number of estimators at which boosting is terminated. In case of perfect fit, the learning procedure is stopped early.
-  Greater the n_estimators value base model will overfit.
-  Smaller the n_estimators value base model will underfit.
-  Hence we need to find optimum value of n_estimators for best results in Adaboost.

### learning_rate=1.0
- Weight applied to each classifier at each boosting iteration
- A higher learning rate increases the contribution of each classifier
- There is a trade-off between the learning_rate and n_estimators parameters
- Values must be in the range (0.0, inf).
- Basically the idea is to slow down learning allowing to dramatically reduce overfittig.
- As we reduces the learning_rate , amplitude of weights.
- learning_rate is also known as Shrinkage parameter.
- With help of Learning rate we can incraese n_estimators and reduces learning_rate to make a perfect balance.

### algorithm{‘SAMME’, ‘SAMME.R’}, default=’SAMME.R’
- Defaul Algorithm is SAMME.R
- If ‘SAMME.R’ then use the SAMME.R real boosting algorithm.
- If ‘SAMME’ then use the SAMME discrete boosting algorithm.
- The SAMME.R algorithm typically converges faster than SAMME, achieving a lower test error with fewer boosting iterations.

### Code 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions 
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
```

```python
np.random.seed(42)
X,y=make_circles(n_samples=500,noise=0.35,factor=0.1,random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
```

```python
abc=AdaBoostClassifier()
#Using Default hyperparameters.
np.mean(cross_val_score(abc,X,y,scoring='accuracy',cv=10))
```

```python
abc.fit(X,y)
```

#### Plotting Decision Boundry

```python
def plot_decision_boundry(clf):
    plt.figure(figsize=(12, 8))
    x_range = np.linspace(X.min(), X.max(), 100)
    xx1, xx2 = np.meshgrid(x_range, x_range)
    y_hat = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])
    y_hat = y_hat.reshape(xx1.shape)
    plt.contourf(xx1, xx2, y_hat, alpha=0.2)
    plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
    plt.title("Adaboost Classifier")
    plt.show()
plot_decision_boundry(abc)
```

- from above graph you can see that there are some mistakes in Decision boundry of graph blue points are in yellow region and yellow points are in blue region.

##### Tunnig the Parameters

```python
abc=AdaBoostClassifier(n_estimators=500)
abc.fit(X,y)
plot_decision_boundry(abc)
```

```python
abc=AdaBoostClassifier(n_estimators=1500)
abc.fit(X,y)
plot_decision_boundry(abc)
```

```python
abc=AdaBoostClassifier(n_estimators=1500,learning_rate=0.1)
abc.fit(X,y)
plot_decision_boundry(abc)
```

- From above graphs you can notice that Graph with n_estimators 500 and 1500 overfits on the data and try to classify each and every point.
- While in graph with learning_rate=0.1 you can see that n_estimator=1500 even though it perform optimum and gives best result due to effect of learning_rate.
- In this way learning_rate may affect and vary the result of estimator significantly. 

#### Parameter Tuning Using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
grid['algorithm'] = ['SAMME', 'SAMME.R']


grid_search = GridSearchCV(estimator=AdaBoostClassifier(), param_grid=grid, n_jobs=-1, cv=10, scoring='accuracy')
# execute the grid search
grid_result = grid_search.fit(X, y)
# summarize the best score and configuration
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

- from above you can see the best parameters for AdaBoostClassifier using GridSearchCv 
- Surprisingly  algorithm is SAMME , learnin_rate=0.1 and n_estimators=500
