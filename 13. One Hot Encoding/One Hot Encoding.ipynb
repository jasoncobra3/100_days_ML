# 3. OneHotEncoding Using Sklearn

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OHE=OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
X_train_new=OHE.fit_transform(X_train[["fuel","owner"]])
X_test_new=OHE.fit_transform(X_test[["fuel",'owner']])
```

# 4.OneHotEncoding With Top Categories

```python
data=pd.read_csv("cars.csv")
```

```python
data["brand"].value_counts()
```

```python
counts=data["brand"].value_counts()
```

```python
data["brand"].nunique()
threshold=100
```

```python
repl=counts[counts<=threshold].index
```

```python
pd.get_dummies(data["brand"].replace(repl,"uncommon"),dtype=int).sample(5)
```
