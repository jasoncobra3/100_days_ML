```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
np.random.seed(42)
X=np.random.rand(100,1)-0.5
y=3*X[:,0]**2 + 0.05 * np.random.randn(100)
```

```python
df=pd.DataFrame()
```

```python
df["X"]=X.reshape(100)
df['y']=y
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.title("X V/s Y")
```

# Prediction of Base model 1

```python
df['pred1']=df['y'].mean()
```

```python
df
```

##  Calculating pseudo Residual of 1st Base Model

```python
df['res1']=df['y']-df['pred1']
```

```python
df
```

```python
plt.scatter(df['X'],df['y'])
plt.plot(df["X"],df["pred1"],color='purple')
```

- from above graph you can see that , for 1st base model as the predicted value is same for all columns it is shown by straight line in above grpah

# Model 2

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree1.fit(df["X"].values.reshape(100,1),df["res1"].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```

- You can see above tree contains 8 leaf nodes

```python
#generating X_test
X_test=np.linspace(-0.5,0.5,500)
```

```python
y_pred = 0.265458 + tree1.predict(X_test.reshape(500, 1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
```

- In the above graph you can see that , now model 2 to try to fit over data by understanding the mistake of 1st base model.
- Also it is understanding the nature of the data

## Making Prediction of Model 2

```python
df['pred2'] = 0.265458 + tree1.predict(df['X'].values.reshape(100,1))
```

```python
df
```

```python
df['res2']=df["y"]-df['pred2']
```

```python
#Understanding the Pseudo Residual of model 2
df
```

# Model 3

```python
tree2=DecisionTreeRegressor(max_leaf_nodes=8)
```

```python
tree2.fit(df["X"].values.reshape(100,1),df['res2'].values)
#res2 is the error of 1st and second model 
#We are training the model with X and res2 column
```

```python
y_pred = 0.265458 + sum(regressor.predict(X_test.reshape(-1, 1)) for regressor in [tree1,tree2])
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
plt.title('X vs y')
```

- From above graph you can understand that , model 3 is performing better than model 2 since it is learning from the mistakes of model 1 and model 2 using res2 column.
- In above graph model 3 is understanding the data better than model 2 tries to cover all points in the data, and making prediction more accurately.

# Note: In above example we are using learning_rate =1 , for predictions and further calculations.

```python
def gradient_boost(X,y,number,lr,count=1,regs=[],foo=None):

  if number == 0:
    return 
  else:
    # do gradient boosting

    if count > 1:
      y = y - regs[-1].predict(X)
    else:
      foo = y

    tree_reg = DecisionTreeRegressor(max_depth=5, random_state=42)
    tree_reg.fit(X, y)

    regs.append(tree_reg)

    x1 = np.linspace(-0.5, 0.5, 500)
    y_pred = sum(lr * regressor.predict(x1.reshape(-1, 1)) for regressor in regs)

    print(number)
    plt.figure()
    plt.plot(x1, y_pred, linewidth=2)
    plt.plot(X[:, 0], foo,"r.")
    plt.show()

    gradient_boost(X,y,number-1,lr,count+1,regs,foo=foo)
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
gradient_boost(X,y,5,lr=1)
```

- From above Graph you can see that as the learning_rate=1 , the models we are training on the data are overfit on it.
- Hence to handle overfitting over data , we can tune the learning_rate=0.5 so that we can move in the correct direction in slow steps.
- You can tune the above function to achieve precise result.
- Thus learning rate plays important role in GradientBoost Algorithm.
