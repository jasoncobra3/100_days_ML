## Random Sample Imputation

### It will not change the Distribution of Data Like Other Imputation Methods

### Useful in Linear and Not useful in Decision Tree Problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sbs
```

```python
df=pd.read_csv("train.csv",usecols=["Age","Fare","Survived"])
```

```python
df.head()
```

### For Numerical Data

```python
df.isnull().mean()*100
```

```python
X=df.drop(columns=["Survived"])
Y=df["Survived"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
X_train["Age_Imputed"]=X_train["Age"]
X_test["Age_Imputed"]=X_test["Age"]
```

```python
X_test.tail()
```

```python
X_train["Age_Imputed"][X_train["Age_Imputed"].isnull()]==X_train["Age"].dropna().sample(X_train["Age"].isnull().sum()).values
X_test["Age_Imputed"][X_test["Age_Imputed"].isnull()]==X_train["Age"].dropna().sample(X_test["Age"].isnull().sum()).values
```

```python
X_train["Age"].dropna().sample(X_train["Age"].isnull().sum()).values
```

```python
X_train["Age"].isnull().sum()
```

```python
X_train
```

```python
sbs.distplot(X_train["Age"],hist=False,label="original")
sbs.distplot(X_train["Age_Imputed"],hist=False,label="imputed")

plt.legend()
plt.show()
```

#### from the above graph we can see that Distribution of "Age" and "Age_Imputed" Column is Same even after imputation

```python
print("Original Variance of Age:",X_train["Age"].var())
print("Variance Of Imputed Age Column:",X_train["Age_Imputed"].var())
```

##### you can see that there is no change in the Variance of columnafter Imputation

```python
X_train[["Age","Fare","Age_Imputed"]].cov()
```

```python
X_train[["Age","Age_Imputed"]].boxplot()
```

##### There is no change or generation of outliers After Imputation in Age Column

### For Categorical Data

```python
data = pd.read_csv('house-train.csv',usecols=['GarageQual','FireplaceQu', 'SalePrice'])
```

```python
data.head()
```

```python
data.isnull().mean()*100
```

```python
X=data
Y=data['SalePrice']
```

```python
X_train,X_test,Y_train ,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train["GarageQual_imputed"]=X_train["GarageQual"]
X_test["GarageQual_imputed"]=X_test["GarageQual"]

X_train["FireplaceQu_imputed"]=X_train["FireplaceQu"]
X_test["FireplaceQu_imputed"]=X_test["FireplaceQu"]
```

```python
X_train.sample(5)
```

```python
X_train["GarageQual_imputed"][X_train["GarageQual_imputed"].isnull()]=X_train["GarageQual"].dropna().sample(X_train["GarageQual"].isnull().sum()).values
X_test["GarageQual_imputed"][X_test["GarageQual_imputed"].isnull()]=X_train["GarageQual"].dropna().sample(X_test["GarageQual"].isnull().sum()).values

X_train["FireplaceQu_imputed"][X_train["FireplaceQu_imputed"].isnull()]=X_train["FireplaceQu"].dropna().sample(X_train["FireplaceQu"].isnull().sum()).values
X_test["FireplaceQu_imputed"][X_test["FireplaceQu_imputed"].isnull()]=X_train["FireplaceQu"].dropna().sample(X_test["FireplaceQu"].isnull().sum()).values
```

```python
temp=pd.concat([
    X_train["GarageQual"].value_counts()/len(X_train["GarageQual"].dropna()),
    X_train["GarageQual_imputed"].value_counts()/len(X_train)
],axis=1)

temp.columns=["original","imputed"]
```

```python
temp
```

```python
temp=pd.concat([
    X_train["FireplaceQu"].value_counts()/len(X_train["FireplaceQu"].dropna()),
    X_train["FireplaceQu_imputed"].value_counts()/len(X_train)
],axis=1)

temp.columns=["original","imputed"]
```

```python
temp
```

```python
for category in X_train["FireplaceQu"].dropna().unique():
    sbs.distplot(X_train[X_train["FireplaceQu"]==category]["SalePrice"],hist=False,label=category)

plt.show()
```

```python
for category in X_train["GarageQual"].dropna().unique():
    sbs.distplot(X_train[X_train["GarageQual"]==category]["SalePrice"],hist=False,label=category)
plt.show()
```

### We can see that Random Sample Impuation for categorical Columns , it is causing changes in distribution of data
