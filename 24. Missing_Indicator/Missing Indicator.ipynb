# Missing_Indicator

### This method will generate Truth Table Automatically for Missinng Value <br> It will write True Value for Null value, create a New Table

```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer , MissingIndicator
```

```python
df=pd.read_csv("train.csv",usecols=["Age","Fare","Survived"])
```

```python
df.head()
```

```python
X=df.drop(columns=["Survived"])
Y=df["Survived"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
si=SimpleImputer()
X_train_trf=si.fit_transform(X_train)
X_test_trf=si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,Y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
mi=MissingIndicator()
mi.fit(X_train)
```

```python
mi.features_
```

```python
X_train_missing=mi.transform(X_train)
```

```python
X_train_missing
```

```python
X_test_missing=mi.transform(X_test)
```

```python
X_test_missing
```

```python
X_train["Age_Na"]=X_train_missing
```

```python
X_train.head()
```

```python
X_test["Age_Na"]=X_test_missing
```

```python
X_test
```

```python
si=SimpleImputer()
X_train_trf1=si.fit_transform(X_train)
X_test_trf1=si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train_trf1,Y_train)

pred2=clf.predict(X_test_trf1)

accuracy_score(pred2,Y_test)
```

```python
si=SimpleImputer(add_indicator=True)
#We can us Missing Indicator in Simple Imputer by Passing The Add_indicator Parameter
```

```python
X_train_Trf=si.fit_transform(X_train)
X_test_Trf=si.transform(X_test)
```

```python
clf1=LogisticRegression()

clf1.fit(X_train_Trf,Y_train)
predictions=clf1.predict(X_test_Trf)
```

```python
accuracy_score(predictions,Y_test)
```
